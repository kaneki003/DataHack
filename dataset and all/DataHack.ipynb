## Feature Encoding

## Splitting training data

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(training_data, train_res, test_size=0.4, random_state=1)
train_data=training_data.copy()
```

```python
nan_lst=[]
for col in cols:
    if(train_data[col].isna().sum()!=0):
        nan_lst.append(col)

mod_nan_lst=[]
for col in nan_lst:
    if pd.isna(train_data[col].unique()[0]):
        mod_nan_lst.append(train_data[col].unique()[1])
    else:
        mod_nan_lst.append(train_data[col].unique()[0])

for i in range(len(nan_lst)):
    X_test[nan_lst[i]]=X_test[nan_lst[i]].fillna(mod_nan_lst[i])
    X_train[nan_lst[i]]=X_train[nan_lst[i]].fillna(mod_nan_lst[i])
    train_data[nan_lst[i]]=train_data[nan_lst[i]].fillna(mod_nan_lst[i])
```

```python
from sklearn.preprocessing import OneHotEncoder
one_hot=OneHotEncoder(categories='auto')

nom_col=['race','sex','marital_status','hhs_geo_region','employment_industry','employment_occupation']

for col in nom_col:
    # Fit OneHotEncoder using only the training data
    one_hot.fit(training_data[[col]])
    
    # Transform both training and test data
    train_encoded = one_hot.transform(X_train[[col]])
    test_encoded = one_hot.transform(X_test[[col]])
    
    # Create DataFrames with correct column names
    train_data = pd.DataFrame(train_encoded.toarray(), columns=one_hot.get_feature_names_out())
    test_data = pd.DataFrame(test_encoded.toarray(), columns=one_hot.get_feature_names_out())
    
    # Drop original column and concatenate encoded data
    X_train = X_train.drop([col], axis=1).reset_index(drop=True)
    X_train = pd.concat([X_train, train_data], axis=1)
    
    X_test = X_test.drop([col], axis=1).reset_index(drop=True)
    X_test = pd.concat([X_test, test_data], axis=1)
```

```python
cat_ord={'18 - 34 Years': 0, '35 - 44 Years': 1, '45 - 54 Years': 2,'55 - 64 Years': 3,'65+ Years': 4,
         'College Graduate': 2, 'Some College': 1, '12 Years': 0, 
         '< 12 Years': 3,'Below Poverty': 0, '<= $75,000, Above Poverty': 1, '> $75,000': 2,'Own': 0, 'Rent': 1,
        'Non-MSA': 0, 'MSA, Not Principle  City': 1,'MSA, Principle City': 2,
        'Employed': 0, 'Not in Labor Force': 1, 'Unemployed': 2}

def ord(x):
    return cat_ord[x]

ord_col=['age_group','education','income_poverty','rent_or_own','census_msa','employment_status']
for col in ord_col:
    X_test[col]=X_test[col].apply(ord)
    X_train[col]=X_train[col].apply(ord)
```

## Testing out different models

## Random Forests

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score
import numpy as np

rf_model = RandomForestClassifier()
rf_model.fit(X_train, y_train)

# Predict probabilities for each class
ytrain_pred = rf_model.predict_proba(X_train)
ytest_pred = rf_model.predict_proba(X_test)

# Reshape predictions to 2D array
ytrain_pred = np.array([pred[:, 1] for pred in ytrain_pred]).T
ytest_pred = np.array([pred[:, 1] for pred in ytest_pred]).T

# Calculate ROC AUC for each label and take the average
train_roc_auc = roc_auc_score(y_train, ytrain_pred, multi_class='ovo', average='macro')
test_roc_auc = roc_auc_score(y_test, ytest_pred, multi_class='ovo', average='macro')

print(f'RF train roc-auc: {train_roc_auc}')
print(f'RF test roc-auc: {test_roc_auc}')
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.multiclass import OneVsRestClassifier
from sklearn.preprocessing import StandardScaler

log_classifier=OneVsRestClassifier(LogisticRegression(max_iter=1000))
log_classifier.fit(X_train, y_train)

ytrain_pred = log_classifier.predict_proba(X_train)
ytest_pred = log_classifier.predict_proba(X_test)

train_roc_auc = roc_auc_score(y_train, ytrain_pred, multi_class='ovr', average='macro')
test_roc_auc = roc_auc_score(y_test, ytest_pred, multi_class='ovr', average='macro')

print(f'RF train roc-auc: {train_roc_auc}')
print(f'RF test roc-auc: {test_roc_auc}')
```

## Adaboost Classifier

```python
from sklearn.ensemble import AdaBoostClassifier

ada_classifier=OneVsRestClassifier(AdaBoostClassifier())
ada_classifier.fit(X_train, y_train)

# Predict probabilities for each class
ytrain_pred = ada_classifier.predict_proba(X_train)
ytest_pred = ada_classifier.predict_proba(X_test)

# Calculate ROC AUC for each label and take the average
train_roc_auc = roc_auc_score(y_train, ytrain_pred, multi_class='ovr', average='macro')
test_roc_auc = roc_auc_score(y_test, ytest_pred, multi_class='ovr', average='macro')

print(f'RF train roc-auc: {train_roc_auc}')
print(f'RF test roc-auc: {test_roc_auc}')
```
